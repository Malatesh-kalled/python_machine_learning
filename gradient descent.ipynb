```python
import numpy as np
import matplotlib.pyplot as plt

def loss_function(x):
    return x**2 + 3*x + 2

def gradient(x):
    return 3*x + 4

def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    path = [(x, loss_function(x))]

    for i in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        path.append((x, loss_function(x)))

    return x, path

initial_x = 0.0
learning_rate = 0.1
num_iterations = 100

final_x, path = gradient_descent(initial_x, learning_rate, num_iterations)

print("Gradient Descent Finished")
print(f"Final x = {final_x}, Final Loss = {loss_function(final_x)}")

x_values, loss_values = zip(*path)

plt.figure(figsize=(10, 6))
plt.plot(x_values, loss_values, '-o')
plt.xlabel('x')
plt.ylabel('Loss')
plt.title('Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
