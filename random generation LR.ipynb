```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

random.seed(42)
num_points = 10
x = np.array([random.uniform(0, 10) for _ in range(num_points)]).reshape((-1, 1))
y = np.array([2 * xi + 3 + random.uniform(-1, 1) for xi in x]).reshape((-1, 1))

plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)
mse = mean_squared_error(y, pred)

plt.scatter(x, y)
plt.plot(x, pred, color='blue', marker='*')
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```
