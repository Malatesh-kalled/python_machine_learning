```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,3,5,6,20,13]).reshape((-1,1))
y=np.array([4,8,5,16,21,26])
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Blue',marker='*')
print("MSE:",mse)
print("intercept:",SLR.intercept_)

print("slope:",SLR.coef_)
```
