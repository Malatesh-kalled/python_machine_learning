```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,3,5,6,20,13]).reshape((-1,1))
y=np.array([4,8,5,16,21,26])
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Blue',marker='*')
print("MSE:",mse)
print("intercept:",SLR.intercept_)

print("slope:",SLR.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

random.seed(42)
num_points = 10
x = np.array([random.uniform(0, 10) for _ in range(num_points)]).reshape((-1, 1))
y = np.array([2 * xi + 3 + random.uniform(-1, 1) for xi in x]).reshape((-1, 1))

plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)
mse = mean_squared_error(y, pred)

plt.scatter(x, y)
plt.plot(x, pred, color='blue', marker='*')
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_excel('/content/drive/MyDrive/data.xlsx')
x = data['x'].values.reshape(-1, 1)
y = data['y'].values

model = LinearRegression()
model.fit(x, y)

slope = model.coef_[0]
intercept = model.intercept_

plt.scatter(x, y)
plt.plot(x, model.predict(x), color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print("Slope:", slope)
print("Intercept:", intercept)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_excel('/content/drive/MyDrive/data.xlsx')

X = data[['x1', 'x2', 'x3']].values
y = data['y'].values

model = LinearRegression()
model.fit(X, y)

coefficients = model.coef_
intercept = model.intercept_


print("Coefficients:", coefficients)
print("Intercept:", intercept)


predictions = model.predict(X)


plt.scatter(y, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs. Predictions')
plt.show()
```
